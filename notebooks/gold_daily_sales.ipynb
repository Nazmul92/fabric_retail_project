```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
spark.sql("""
CREATE OR REPLACE TABLE gold.daily_sales AS
SELECT 
    transaction_date,
    SUM(total_amount) AS daily_total_sales
FROM 
    silver.orders
GROUP BY 
    transaction_date
""")
```

```python
# Verify the data in the Gold table
spark.sql("SELECT * FROM daily_sales LIMIT 10").show()
```

```python
spark.sql("""
CREATE OR REPLACE TABLE gold.monthly_sales AS
SELECT
    DATE_TRUNC('month', CAST(transaction_date AS DATE)) AS sales_month,
    SUM(total_amount) AS monthly_total_sales
FROM silver.orders
GROUP BY DATE_TRUNC('month', CAST(transaction_date AS DATE))
""")
```

```python
spark.sql("SELECT * FROM gold.monthly_sales ORDER BY sales_month LIMIT 10").show(truncate=False)
```

```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
spark.sql("""
CREATE OR REPLACE TABLE gold.category_sales AS
SELECT 
    p.category AS product_category,
    SUM(o.total_amount) AS category_total_sales
FROM 
    silver.orders o
JOIN 
    silver.products p ON o.product_id = p.product_id
GROUP BY 
    p.category
""")
```

```python
# Verify the data in the Gold table
spark.sql("SELECT * FROM gold.category_sales LIMIT 10").show()
```
